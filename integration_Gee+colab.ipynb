```python
import ee
import geemap

# Authenticate with Google
ee.Authenticate()
ee.Initialize(project='ee-2303c50091')
```

```python
# Load the required dataset
districts = ee.FeatureCollection('FAO/GAUL/2015/level2')

# Filter for Warangal district
warangal = districts.filter(ee.Filter.eq('ADM2_NAME', 'Warangal'))

# Create a map
Map = geemap.Map()

# Add Warangal district boundary
Map.addLayer(warangal, {}, 'Warangal District')

# Center the map on Warangal
Map.centerObject(warangal, zoom=8)

# Display the map
Map
```
